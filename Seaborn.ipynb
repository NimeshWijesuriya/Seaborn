```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
data=sns.load_dataset('car_crashes')
```

```python
data.head()
```

```python
#Distribution Plot
```

```python
sns.displot(data['not_distracted'],kde=True)
```

```python
#Joint Plot
```

```python
sns.jointplot(x='speeding',y='alcohol',data=data,kind='hex')
```

```python
#KDE Plot
```

```python
sns.kdeplot(data['alcohol'])
```

```python
#Pair Plot
```

```python
sns.pairplot(data)
```

```python
#Rug plot
```

```python
sns.rugplot(data['speeding'])
```

```python
#Styling
```

```python
sns.set_style('dark')
plt.figure(figsize=(8,4))
sns.set_context('talk',font_scale=1)
sns.jointplot(x='speeding',y='alcohol',data=data,kind='reg')
```

# Categorical Plots

```python
#Bar Plot
```

```python
data=sns.load_dataset('tips')
```

```python
sns.barplot(x='sex',y='total_bill',data=data,estimator=np.std)
```

```python
#Count Plot
```

```python
sns.countplot(x='sex',data=data)
```

```python
#Box Plot
```

```python
sns.boxplot(x='day',y='total_bill',data=data,hue='sex')
plt.legend(loc=0)
```

```python
#Violin Plot
```

```python
sns.violinplot(x='day',y='total_bill',data=data,hue='sex')
plt.legend(loc=0)
```

```python
#Strip Plot
```

```python
plt.figure(figsize=(8,5))
sns.stripplot(x='day',y='total_bill',data=data,hue='sex')
```

```python
#Swarm Plot
```

```python
sns.violinplot(x='day',y='total_bill',data=data)
sns.swarmplot(x='day',y='total_bill',data=data,color='red')
```

```python
#Palettes
```

```python
plt.figure(figsize=(8,5))
sns.set_style('white')
sns.set_context('talk')
sns.stripplot(x='day',y='total_bill',data=data,hue='sex',palette='magma')
```
